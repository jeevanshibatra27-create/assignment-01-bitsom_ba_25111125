```python
sentences = [
    # Cricket
    "The batsman scored a century in the final match.",
    "The bowler took five wickets in the tournament.",
    "Cricket fans watched the thrilling game.",

    # Cooking
    "I love baking chocolate chip cookies.",
    "Cooking pasta requires boiling water first.",
    "Grilling vegetables adds a smoky flavor.",

    # Cybersecurity
    "Phishing attacks can steal personal information.",
    "Strong passwords prevent unauthorized access.",
    "Ransomware can lock all files on a computer.",
    "Cybersecurity experts monitor network traffic for threats."
]
```

```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
model = SentenceTransformer('all-MiniLM-L6-v2')
embeddings = model.encode(sentences)
```

```python
sim_matrix = cosine_similarity(embeddings)
print(sim_matrix)
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(sim_matrix, annot=True, fmt=".2f", cmap="coolwarm", xticklabels=sentences, yticklabels=sentences)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.show()
```

```python
query = "The bowler took three wickets in one over"
query_emb = model.encode([query])
cos_sim = cosine_similarity(query_emb, embeddings)[0]

top_indices = cos_sim.argsort()[-2:][::-1]  # top 2
for idx in top_indices:
    print(f"Sentence: {sentences[idx]}")
    print(f"Similarity Score: {cos_sim[idx]:.2f}\n")
```
